```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm 
from sklearn.metrics import accuracy_score
```

#### Loading and Analyzing Data

```python
diabetes_data=pd.read_csv("diabetes.csv")
```

```python
diabetes_data.head()
```

```python
diabetes_data.describe()
```

```python
diabetes_data.shape
```

```python
diabetes_data['Outcome'].value_counts()
```

##### 0-----> Non-Diabetic
##### 1-----> Diabetic

#### Extracting Feature

```python
X=diabetes_data.drop(columns='Outcome',axis=1)
Y=diabetes_data['Outcome']
```

### Data Standarization

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data=scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X=standarized_data
```

### Train Test Split

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=5)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Model Training

```python
classifire=svm.SVC(kernel='linear')
```

```python
classifire.fit(X_train,y_train)
```

### Evaluation

#### Accuracy Score

```python
# tesing accuracy on training data
X_train_prediction=classifire.predict(X_train)
X_train_accuracy=accuracy_score(X_train_prediction,y_train)
print("Accuracy on Training Data: ",X_train_accuracy)
```

```python
#testing accuracy on testing data
X_test_prediction=classifire.predict(X_test)
X_test_accuracy=accuracy_score(X_test_prediction,y_test)
print("Accuracy on Testing Data: ",X_test_accuracy)
```

### Results
   - Training Accuracy: 79.64%
   - Test Accuracy: 75.32%
   
   The small gap (~4%) between training and test accuracy shows the model 
   generalizes well. SVM with StandardScaler performs better than Logistic 
   Regression without scaling for this type of medical data.

### Making a predictive System

```python
input_data=(1,85,66,29,0,26.6,0.351,31)

#converting input to numpy array
input_array=np.asarray(input_data)

#reshaping according to label columns  
input_array_reshaped=input_array.reshape(1,-1)

scaled_input=scaler.transform(input_array_reshaped)

prediction=classifire.predict(scaled_input)
if prediction[0]==0: 
    print("Patient is NOT Diabetic")
else:
    print("Patient is Diabetic")
```
